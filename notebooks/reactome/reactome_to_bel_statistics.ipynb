# Reactome to BEL Statistics

### This notebook generates the statisitics for Reactome entity and interaction types captured in their transformation into BEL nodes and edges, respectively.

```python
from pathme.constants import REACTOME_BEL, BEL_STATS_COLUMN_NAMES
from pathme.utils import get_files_in_folder, get_bel_stats, get_bel_types
```

### Generate a dataframe of pathways from Reactome converted into BEL.

```python
reactome_to_bel_statistics = get_bel_stats(REACTOME_BEL)
```

```python
reactome_to_bel_statistics
```

### Plot statistics for all entity and interaction types in BEL graphs generated from Reactome.

```python
bel_statistics = reactome_to_bel_statistics.sum(axis=0)

reactome_to_bel = bel_statistics.plot(kind='bar', figsize=(15,10), color='#df3f18')

reactome_to_bel.title.set_size(20)
reactome_to_bel.set_title('Reactome to BEL Statistics', fontdict={'size': 22})
reactome_to_bel.set_xlabel('BEL Type', fontdict={'size': 18})
reactome_to_bel.set_ylabel('Frequency', fontdict={'size': 18})
```
