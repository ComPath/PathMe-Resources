# Notch Signaling Pathway Example

## Example conversion of the KEGG Pathways 'Notch Signaling Pathway' KGML file into BEL

```python
import matplotlib.pyplot as plt
import pandas as pd
import os

from pathme.kegg import utils
from pathme.kegg.convert_to_bel import kegg_to_bel
from pathme.constants import KEGG_FILES
from pathme.kegg.utils import get_kegg_pathway_ids, get_kegg_statistics
from pathme.utils import get_files_in_folder
from pathme_viewer.manager import Manager

from pybel_tools import summary

from pybel_jupyter import to_jupyter

from bio2bel_chebi import Manager as ChebiManager
from bio2bel_hgnc import Manager as HgncManager
```

```python
# Initiate HGNC Manager
hgnc_manager = HgncManager()

# Initiate ChEBI Manager
chebi_manager = ChebiManager()

# Initiate PathMe-Viewer Manager
manager = Manager.from_connection()
```

```python
notch_pathway = os.path.join(KEGG_FILES, 'hsa04330.xml')
```

### Convert the Notch Signaling Pathway from KEGG to BEL in un-flattened mode

```python
notch_graph = kegg_to_bel(notch_pathway, hgnc_manager, chebi_manager, flatten=False)
```

### Generate entity and interaction type statistics for the notch pathway in un-flattened mode

```python
summary.plot_summary(notch_graph, plt)
plt.savefig('notch_statistics_unflattened')
```

### Visualize the Notch Signaling Pathway as a BEL Graph


```python
to_jupyter(notch_graph)
```

### Convert the Notch Signaling Pathway from KEGG to BEL with flattened mode enable

```python
flat_notch_graph = kegg_to_bel(notch_pathway, hgnc_manager, chebi_manager, flatten=True)
```

### Generate entity and interaction type statistics for the notch pathway in flattened mode

```python
summary.plot_summary(flat_notch_graph, plt)
plt.savefig('notch_statistics_flattened')
```

### Visualize the Notch Signaling Pathway as a BEL Graph

```python
to_jupyter(flat_notch_graph)
```
